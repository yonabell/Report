```
import pandas as pd
```

```
df = 'Quarterly comparision_2.xlsx'
```

```
# Read the CSV file content, using the first row as the header
df = pd.read_excel(df, header=0)
```

```
# Function to convert dates to quarters
def date_to_quarter(date_str):
    date = pd.to_datetime(date_str, dayfirst=True)
    quarter = (date.month - 1) // 3 + 1  # Get quarter
    return f"Q{quarter}/{date.year}"  # Format as Qn/yyyy
```

```
# Apply the conversion function
df['Quarter'] = df['Date'].apply(date_to_quarter)
```

```
# Drop the original date column
df = df.drop(columns=['Date'])
```

```
df.shape
```

```
# Move 'Quarter' column to the first position
cols = list(df.columns)
cols.insert(0, cols.pop(cols.index('Quarter')))
df = df[cols]
```

```
# Ensure 'Quarter' column exists and remove non-numeric columns from max calculation
numeric_columns = df.columns[1:]  # Assuming the first column is 'Quarter', and rest are numeric
```

```
df
```

```
# Filter the rows based on the 'Quarter' column
q1_rows = df[df['Quarter'].str.contains("Q1")]
q2_rows = df[df['Quarter'].str.contains("Q2")]
q3_rows = df[df['Quarter'].str.contains("Q3")]
q4_rows = df[df['Quarter'].str.contains("Q4")]
```

```
df
```

```
# Calculate the maximum values for each category in each quarter
max_q1 = q1_rows[numeric_columns].max(axis=0)
max_q2 = q2_rows[numeric_columns].max(axis=0)
max_q3 = q3_rows[numeric_columns].max(axis=0)
max_q4 = q4_rows[numeric_columns].max(axis=0)
```

```
df
```

```
# Combine the max values into a DataFrame
max_values = pd.DataFrame({
    'Max_Q1': max_q1.values,
    'Max_Q2': max_q2.values,
    'Max_Q3': max_q3.values,
    'Max_Q4': max_q4.values
}, index=numeric_columns)
```

```
# Transpose the result so that Max_Q1, Max_Q2, etc. are rows, and categories are columns
max_values_transposed = max_values.T

# Output the result
print(max_values_transposed)
```
